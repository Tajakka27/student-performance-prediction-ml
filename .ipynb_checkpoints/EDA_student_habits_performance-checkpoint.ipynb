## ðŸ“Š Univariate Analysis: Visualization Guide


### ðŸ”  For **Categorical Values**, we usually use:
- **Bar Chart / Countplot** â€“ to display frequency of each category.
- **Pie Chart** â€“ to show proportions (used less often in data analysis).
- **Stacked Bar Chart** â€“ to compare distributions between two categorical variables.
- **Box/Violin Plots** (if comparing against a numeric variable) â€“ to show spread of numeric data across categories.


### ðŸ”¢ For **Numerical Values**, we usually use:
- **Histogram** â€“ to show the distribution of values.
- **Box Plot** â€“ to detect outliers and see the spread.
- **Violin Plot** â€“ to visualize the distribution and density.
- **CDF/PDF Plots** â€“ to observe cumulative or probability distribution patterns.

# Why is a Box Plot Important?

<p>
A <strong>box plot</strong> (or box-and-whisker plot) is a powerful visualization tool used to understand the distribution of numerical data. It is particularly useful for:
</p>

<ul>
  <li> <strong>Identifying Outliers:</strong> Points that lie outside the whiskers (typically 1.5Ã—IQR from Q1 or Q3) are outliers â€” values that may be errors or unusual cases worth further investigation.</li>
  <li> <strong>Understanding Distribution:</strong> It shows the <em>median</em>, <em>quartiles</em>, and <em>spread</em> of the data at a glance. This helps understand if the data is symmetric, skewed, or contains any concentration in specific ranges.</li>
  <li> <strong>Comparing Multiple Variables:</strong> When used across multiple features, box plots make it easy to compare variability and central tendency side-by-side.</li>
  <li><strong>Detecting Skewness:</strong> If the median is not centered in the box or one whisker is longer, it suggests skewness in the data.</li>
</ul>

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

<center><h1> Step 5: Bivariate Analysis <h2></center>



### Numerical vs Numerical:
Use **scatter plots**, **correlation heatmaps**, or **joint plots** to explore the relationship between two numerical variables.

### Numerical vs Categorical:
Use **box plots**, **violin plots**, **bar plots**, or **point plots** to examine the distribution and comparison of a numerical variable across different categories of a categorical variable.

### Categorical vs Categorical:
Use **stacked bar charts** or **grouped bar charts** to visualize the relationship between two categorical variables by showing counts or proportions.

### When You Need to Observe Relationship Using Probability Distribution:

**Joint Plot**: Combines scatter plot, histogram, and kernel density estimation (KDE) to show relationships between two variables. It helps to visualize the relationship between numerical variables while showing their individual distributions.

**Pair Plot**: Displays scatter plots and histograms for every pair of variables in the dataset, allowing you to observe relationships for multiple variable combinations at once. It is a useful tool to examine how different variables interact and understand overall data patterns.

<center><h2> Step 7: Data Preprocessing </h2></center>

<h2> Encoding Categorical Variables</h2>

<p>
To prepare the dataset for machine learning models, we performed encoding on the categorical variables. Machine learning algorithms typically require numerical input, so converting categorical variables into a numerical format is a necessary preprocessing step.
</p>

<h3> Label Encoding</h3>
<p>
We applied <strong>Label Encoding</strong> to ordinal categorical features where the categories have an inherent order:
</p>
<ul>
  <li><code>diet_quality</code></li>
  <li><code>parental_education_level</code></li>
  <li><code>internet_quality</code></li>
</ul>
<p>
Label Encoding converts each category into a corresponding integer value. For example, "Poor", "Fair", and "Good" may be converted to 0, 1, and 2 respectively.
</p>

<h3> One-Hot Encoding</h3>
<p>
For nominal categorical features that have no ordinal relationship, we used <strong>One-Hot Encoding</strong>:
</p>
<ul>
  <li><code>gender</code></li>
  <li><code>part_time_job</code></li>
  <li><code>extracurricular_participation</code></li>
</ul>
<p>
This process creates binary columns for each category, avoiding any ordinal misinterpretation. We used 
</p>



```python
from sklearn.preprocessing import LabelEncoder

# Make a copy to avoid modifying original
df_encoded = df.copy()

# Label Encoding
label_cols = ['diet_quality', 'parental_education_level', 'internet_quality']
le = LabelEncoder()
for col in label_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# One-Hot Encoding
one_hot_cols = ['gender', 'part_time_job', 'extracurricular_participation']
df_encoded = pd.get_dummies(df_encoded, columns=one_hot_cols, drop_first=True)

# Display the first few rows of the encoded dataframe
df_encoded.head()
```

# Seperating the X and Y feature

```python
X = df_encoded.drop('exam_score', axis=1)
y = df_encoded['exam_score']
print(X.shape, y.shape) 
```

<h2> Feature Scaling: Standardization with StandardScaler</h2>

<p>
Feature scaling is a crucial step in data preprocessing, especially for machine learning algorithms that are sensitive to the scale of input features (e.g., Logistic Regression, KNN, SVM).
</p>

<p>
In this step, we applied <strong>StandardScaler</strong> from <code>sklearn.preprocessing</code> to normalize the dataset. StandardScaler standardizes features by removing the mean and scaling to unit variance (i.e., it transforms the data to have zero mean and a standard deviation of one).
</p>

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled.shape  # this will show the shape after scaling
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

<h2> Model Training and Evaluation: Linear Regression</h2>

<p>
In this step, we trained a <strong>Linear Regression</strong> model to predict the target variable using the training dataset. Linear Regression is a fundamental algorithm used for modeling the relationship between one or more independent variables and a continuous dependent variable.
</p>

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
model.score(X_test, y_test)
```

# Trying other models to see which model performs the best

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Try multiple models
models = {
    "Linear": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.1),
    "Random Forest": RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
}

regression_metrics = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    regression_metrics[name] = {'RÂ²': r2, 'RMSE': rmse}
    
metrics_df = pd.DataFrame(regression_metrics).T

# Display the DataFrame
print(metrics_df)
```

```python
metrics_df.plot(kind='bar', figsize=(10, 6), color=['Red', 'Blue'])
plt.title('Regression Model Comparison (RÂ² and RMSE)')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

### Model Evaluation: RÂ² (R-squared) and RMSE (Root Mean Squared Error)

To determine which model is the best, I focused on both **RÂ² (R-squared)** and **RMSE (Root Mean Squared Error)**, which are both important metrics for evaluating regression models:

- **RÂ² (R-squared)**: This indicates how well the model's predictions match the actual data. A value closer to 1 suggests a better fit. From your results:
  - The **Lasso** model has the highest RÂ² (0.897386), indicating it explains the most variance in the data.
  
- **RMSE (Root Mean Squared Error)**: This measures the average magnitude of the errors between predicted and actual values. A lower RMSE indicates better performance, as it means the modelâ€™s predictions are closer to the true values. From your results:
  - **Lasso** also has the lowest RMSE (5.129642), meaning it makes the smallest prediction errors.

### Conclusion:
**Lasso** is the best model here because:
- It has the highest RÂ² (indicating it explains the most variance in the data).
- It has the lowest RMSE (indicating it makes the smallest prediction errors).
